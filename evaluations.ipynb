```python
df=pd.read_csv('dataset.csv')
df.head()
```

```python
df.columns=['Time','Immobilized bus','B_Truck','Vehicle_ex','Acc_victim','R_over','Fire_Veh','OIF','IIDF','Lack_of_el','Fire','Po_flooding','Manifestations','DNT','TR','Sem_off','Int_Semaphore','Slowness_in_traffic']
```

```python
df['Time'] = pd.to_datetime(df['Time'])
df
```

```python
df['Hour']=df.Time.dt.hour
df
```

```python
def convert(hour):
    if ((hour > 4) & (hour< 12)):
       return('Morning')
    elif ((hour >= 12) & (hour < 16 )):
       return('Noon')
    elif ((hour >= 16) & (hour < 20 )):
       return('Afternoon')
    elif ((hour >= 20) & (hour < 24 )):
       return('Night')
    elif ((hour > 24) & (hour <= 4 )):
       return('Night')
```

```python
df.Hour=df['Hour'].apply(lambda x: convert(x))
```

```python
df
```

```python
df.Hour.unique()
```

```python
df_dummies=pd.get_dummies(df.Hour)
```

```python
df_dummies
```

```python
merged=pd.concat([df,df_dummies], axis=1)
```

```python
merged
```

```python
df=merged.drop(['Hour','Afternoon','Time'], axis=1)
df.head()
```

```python
X=df.drop('Slowness_in_traffic', axis=1)
y=df[['Slowness_in_traffic']]
```

```python
cols=X.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
X=scaler.fit_transform(X)
X=pd.DataFrame(X, columns=cols)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,y, test_size=0.3)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

# mean squared error and r-2 score

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(Y_test,y_pred)
r2 = r2_score(Y_test,y_pred)
print('mean sqaure error:',mse)
print('R-Sqaured error:',r2)
```

# median absolute error

```python
from sklearn.metrics import median_absolute_error
median_absolute_error(Y_test,y_pred)
```

# mean absolute error

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Y_test,y_pred)
```

# Mean absolute percentage error

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(Y_test,y_pred)
```
